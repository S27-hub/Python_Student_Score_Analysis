```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Students_scores_dataset.csv")
print(df.head())
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df = df.drop("Unnamed: 0",axis = 1)
print(df.head())
```

# Gender Distribution


```python
plt.figure(figsize=(6,5))
ax = sns.countplot(data = df, x = "Gender")
ax.bar_label(ax.containers[0])
plt.show
```

```python
# from the above chart we can observe that the number of females is more than the number of males.
```

```python
gg = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gg)
```

```python
plt.figure(figsize=(6,6))
sns.heatmap(gg, annot = True )
plt.title("Relationship between Parent's Education and students score")
plt.show
```

```python
# Analysis: Higher the education the higer impact it makes on the marks of the students
```

```python
gg1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gg1)
```

```python
sns.heatmap(gg1, annot = True )
plt.title("Relationship between Parent's Marital Status and students score")
plt.show
```

```python
# the marital status of the parents have little to no impact on the childeren score.
```

# Distribution of Ethnic group

```python
groupA = df.loc[(df['EthnicGroup']=="group A")].count()
print(groupA)
```

```python
groupA = df.loc[(df['EthnicGroup']=="group A")].count()
groupB = df.loc[(df['EthnicGroup']=="group B")].count()
groupC = df.loc[(df['EthnicGroup']=="group C")].count()
groupD = df.loc[(df['EthnicGroup']=="group D")].count()
groupE = df.loc[(df['EthnicGroup']=="group E")].count()

l = ["Group A", "Group B","Group C","Group D","Group E"]
mlist = [groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]

plt.pie(mlist, labels = l, autopct = "%1.2f%%")
plt.title("Distribution of ethnic groups")
plt.show()
```
